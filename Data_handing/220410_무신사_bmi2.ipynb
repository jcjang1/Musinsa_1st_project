```python
full_data = pd.read_csv('/content/데이터합산.csv')
full_data_use = pd.read_csv('/content/데이터합산.csv')
```

```python
use_data = full_data_use[['상품번호', '브랜드명', '성별', '키', '몸무게', '구매한 사이즈', '대분류']]
```

```python
use_data = use_data[use_data['성별'].isin(['남성', '여성'])].reset_index()
```

```python
make_data1 = []
for cm in use_data['키']:
  make_data1.append(int(cm.replace('cm', '')))

use_data['키'] = make_data1
```

```python
make_data2 = []
for kg in use_data['몸무게']:
  make_data2.append(int(kg.replace('kg', '')))

use_data['몸무게'] = make_data2
```

```python
use_data['BMI'] = round(use_data['몸무게'] / ((use_data['키'] / 100) ** 2), 1)
```

```python
bmi_data = []
for bmi in use_data['BMI']:
  if bmi < 18.5:
    bmi_data.append('저체중')
  elif bmi >= 18.6 and bmi < 22.9:
    bmi_data.append('표준')
  elif bmi >= 23.0 and bmi < 24.9:
    bmi_data.append('과체중')
  elif bmi >= 25.0 and bmi < 29.9:
    bmi_data.append('중도비만')
  else:
    bmi_data.append('고도비만')

use_data['정상범위'] = bmi_data
use_data
```

```python
test = re.findall('XL', use_data['구매한 사이즈'][0])
'XL' in test
```

```python
size_data = []

for size in use_data['구매한 사이즈']:
  if 'S' in re.findall('S', size):
    size_data.append('S')
  elif 'M' in re.findall('M', size):
    size_data.append('M')
  elif 'XXL' in re.findall('XXL', size):
    size_data.append('XXL')  
  elif 'XL' in re.findall('XL', size):
    size_data.append('XL')
  elif 'L' in re.findall('L', size):
    size_data.append('L')
  elif 'Free' in re.findall('Free', size):
    size_data.append('Free')
  elif 'F' in re.findall('F', size):
    size_data.append('Free')
  else:
    size_data.append(size)
```

```python
use_data['size'] = size_data
use_data
```
