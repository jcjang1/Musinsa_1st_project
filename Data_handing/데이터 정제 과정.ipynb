# 1월1일부터 3월31일까지를 제외한 데이터 제거

```python
index1 = df_final[df_final['구매날짜']=='2021.12.31'].index
df = df_final.drop(index1)
```

```python
del_lists = ['분', '시간', '1일', '2일', '3일', '4일', '5일', '6일', 
             '2021.', '2020.', '2019.', '2018.', '2017.', '2016.', '2015.', '2014.', '2013.', '2012.', '2011.', '2010.']

for del_list in del_lists:
  df = df[~df['구매날짜'].str.contains(del_list)]
```

```python
df = df.reset_index()
df = df.drop('index', axis = 1)
df.index = df.index+1
df
```
