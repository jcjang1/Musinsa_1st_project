- 키, 몸무게, 구매한 사이즈, 성별
- BMI 계산식 : 몸무게 kg / 키 m 의 제곱
  - 저 ; 18.5 미만
  - 표준 : 18.6 ~ 22.9
  - 과체중 :23.0~24.9
  - 중도비만 : 25.0 ~ 29.9
  - 고도비만 : 30.0 이상

```python
full_data = pd.read_csv('/content/데이터합산.csv')
full_data_use = pd.read_csv('/content/데이터합산.csv')
```

```python
use_data = full_data_use[['상품번호', '브랜드명', '성별', '키', '몸무게', '구매한 사이즈', '대분류']]
```

```python
use_data = use_data[use_data['성별'].isin(['남성', '여성'])].reset_index()
```

```python
make_data1 = []
for cm in use_data['키']:
  make_data1.append(int(cm.replace('cm', '')))

use_data['키'] = make_data1
```

```python
make_data2 = []
for kg in use_data['몸무게']:
  make_data2.append(int(kg.replace('kg', '')))

use_data['몸무게'] = make_data2
```

```python
use_data['BMI'] = round(use_data['몸무게'] / ((use_data['키'] / 100) ** 2), 1)
```

```python
bmi_data = []
for bmi in use_data['BMI']:
  if bmi < 18.5:
    bmi_data.append('저체중')
  elif bmi >= 18.6 and bmi < 22.9:
    bmi_data.append('표준')
  elif bmi >= 23.0 and bmi < 24.9:
    bmi_data.append('과체중')
  elif bmi >= 25.0 and bmi < 29.9:
    bmi_data.append('중도비만')
  else:
    bmi_data.append('고도비만')

use_data['정상범위'] = bmi_data
use_data
```
